```python
from jqdatasdk import *
```

```python
import pandas as pd
```

## 1. get data

```python
stock_code = "000001.XSHE"
l_start_date='2015-01-01'
l_end_date='2022-06-01'
```

### 1.1 get basic data

```python
# get the data, open,close,high,low,volume
df1 = get_price(stock_code, start_date=l_start_date, end_date=l_end_date, frequency='daily')
df1 = df1[["open","close","high","low","volume"]]
df1["change"] = (df1["open"]-df1["close"])/df1["close"]
df1["amplitude"] = (df1["high"]-df1["low"])/df1["low"]
```

```python
df1
```

### 1.2 get factor data

```python
# import 
from jqdatasdk import get_factor_values
```

```python
factor_list = ["Variance20","VR","BIAS5","MAC5","beta","residual_volatility"]
```

```python
# get the factor data
factor_data = get_factor_values(securities=['000001.XSHE'], factors=["Variance20","VR","BIAS5","MAC5","beta","residual_volatility"], 
                                start_date=l_start_date, end_date=l_end_date)
```

```python
# test
# print(factor_data["VR"]["000001.XSHE"].tolist())
factor_data
```

```python
df_factor = factor_data["Variance20"]
df_factor.columns = ["Variance20"]
for factor_name in factor_list[1:]:
    df_tmp = factor_data[factor_name]
    df_tmp.columns = [factor_name]
    df_factor = pd.merge(df_factor, df_tmp, left_index=True, right_index=True)
df_factor
```

### 1.3 merge

```python
df_merge = pd.merge(df1,df_factor,left_index=True, right_index=True)
df_merge
```

```python
# save
df_merge.to_pickle("data.pkl")  
```
