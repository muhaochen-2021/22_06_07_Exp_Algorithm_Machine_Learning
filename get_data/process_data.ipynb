```python
# generate y1 and y2
df_add_y = df.copy()
df_add_y['y_close_f1'] = df_add_y['close'].shift(-1)
df_add_y['y_change_f1'] = df_add_y['change'].shift(-1)
# delete the last row
df_add_y = df_add_y.iloc[:-1,:]
```
